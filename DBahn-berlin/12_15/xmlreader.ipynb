```python
import xml.etree.ElementTree as ET
from datetime import datetime
from zoneinfo import ZoneInfo

BERLIN = ZoneInfo("Europe/Berlin")

def ct_to_datetime(ct: str):
    # "2509021817" -> 2025-09-02 18:17 (Format: %y%m%d%H%M)
    if not ct:
        return None
    return datetime.strptime(ct, "%y%m%d%H%M").replace(tzinfo=BERLIN)

def extract_ar_dp_ct(xml_path: str):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    rows = []
    for s in root.findall("./s"):
        sid = s.get("id")

        ar = s.find("./ar")
        dp = s.find("./dp")

        ar_ct = ar.get("ct") if ar is not None else None
        dp_ct = dp.get("ct") if dp is not None else None

        rows.append({
            "s_id": sid,
            "ar_ct": ar_ct,
            "dp_ct": dp_ct,
            "ar_dt": ct_to_datetime(ar_ct),
            "dp_dt": ct_to_datetime(dp_ct),
        })

    return rows

# Beispiel:
rows = extract_ar_dp_ct("example.xml")
for r in rows:
    print(r["s_id"], r["ar_dt"], r["dp_dt"])
```
