```python
import psycopg2
```

```python
conn = psycopg2.connect(host="localhost", dbname="postgres", user="postgres", password="1234")
cur = conn.cursor()
```

```python
# Run SQL Queries
cur.execute(""" CREATE TABLE IF NOT EXISTS person (
                id INT PRIMARY KEY,
                name VARCHAR(255),
                age INT,
                gender CHAR  
            );
""")


# Apply Changes
conn.commit()

# Close Changes
cur.close()
conn.close()
```
